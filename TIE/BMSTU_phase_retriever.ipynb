```python
import os
import numpy as np
import matplotlib.pyplot as plt
from loader import load_files
from matplotlib.pyplot import figure
from tie import FFTSolver2D
from finite_difference import central_finite_difference
```

```python
i1_path = '../propagation/propagated_fields/bmstu_phase/z=0.200.npy'
i2_path = '../propagation/propagated_fields/bmstu_phase/z=-0.200.npy'
intensities = [np.abs(field) ** 2 for field in load_files([i1_path, i2_path])]

figure(figsize=(10, 8), dpi=100)
plt.imshow(intensities[0], cmap='cividis')
plt.colorbar()
plt.show()
```

```python
dz = 0.0004 # 2 mkm
axial_derivative = central_finite_difference(intensities, dz)
figure(figsize=(10, 8), dpi=100)
plt.imshow(axial_derivative, cmap='cividis')
plt.colorbar()
plt.show()

np.save(f'axial_derivative/bmstu_phase/BMSTU_phase_ad_dz={dz*1000:.3f}mm.npy', axial_derivative)
```

```python
axial_derivative_image = np.load('axial_derivative/bmstu_phase/BMSTU_phase_ad_dz=0.400mm.npy')
figure(figsize=(10, 8), dpi=100)
plt.imshow(axial_derivative_image, cmap='gray')
plt.colorbar()
plt.show()
```

```python
dz = -0.0004  # 2 mkm
px_size = 0.000005  # 5 mkm
wavelength = 0.000000555  # 555 nm
threshold = 0.1

solver = FFTSolver2D(intensities, dz, wavelength, px_size)
retrieved_phase = solver.solve(threshold)
filename = f'BMSTU_phase_dz={dz*1000:.3f}mm.npy'
save_folder = 'retrieved_phases/bmstu_phase'
save_path = os.path.join(save_folder, filename)
np.save(save_path, retrieved_phase)
```

```python
figure(figsize=(10, 8), dpi=100)
plt.imshow(np.load(f'{save_folder}/{filename}'), cmap='gray')
plt.colorbar()
plt.show()
```
